```python
import numpy as np
import pandas as pd
```

```python
all_data = pd.read_csv("/content/dataset assign 4.csv")
all_data.head()
```

Clean up the data


```python
all_data.shape
```

```python
nan_df = all_data[all_data.isna().any(axis = 1)]
display(nan_df.head)

all_data.shape

all_data = all_data.dropna(how = 'all')
all_data.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
print(all_data)
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'] )
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python

all_data['Month'] = pd.to_datetime(all_data['Order Date']).dt.month
all_data.head()
```

```python
def get_city(address):
  return address.split(",")[1].strip(" ")

def get_state(address):
  return address.split(",")[2].strip(" ")[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data.head()
```

1.What was best month for sales and how much was earned in that month?

```python

all_data['Sales'] = all_data['Quantity Ordered'].astype('int')*all_data['Price Each'].astype("float")
all_data.groupby(['Month']).sum()
```

4. What products are most often sold together ?

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

#Referenced: https://stackoverflow.com/questions/27298178/concatenate-strings-from-severa
df['Grouped']= df.groupby('Order ID')['Product']. transform(lambda x: ','.join(x))
df2=df[['Order ID', 'Grouped']].drop_duplicates()
print(df['Grouped'])
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
  row_list = row.split(',')
  count.update(Counter(combinations (row_list, 2)))

for key, value in count.most_common (10): print(key,value)
```
