```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset into a Pandas DataFrame
df = pd.read_csv('/content/sample_data/AQI and Lat Long of Countries2.csv')

# Scatter plot: AQI vs. Lat
plt.scatter(df['lat'], df['AQI Value'])
plt.xlabel('lat')
plt.ylabel('AQI Value ')
plt.title('AQI vs. Latitude')
plt.show()
```

```python
#Scatter Plot of AQI Value and Ozone AQI Value
plt.scatter(df['CO AQI Value'], df['Ozone AQI Value'])
plt.xlabel('CO AQI Value')
plt.ylabel('Ozone AQI Value')
plt.title('CO AQI Value vs. Ozone AQI Value')
plt.show()
```

```python
# Bar plot: Average AQI by country
avg_aqi_by_country = df.groupby('Country')['AQI Value'].mean()
avg_aqi_by_country.plot(kind='bar', figsize=(20, 6))
plt.xlabel('Country')
plt.ylabel('Average AQI')
plt.title('Average AQI by Country')
plt.xticks(rotation=90)
plt.show()
```

```python
# Pie chart: Proportion of AQI categories
aqi_categories = ['Good', 'Moderate', 'Unhealthy', 'Very Unhealthy', 'Hazardous','Unhealthy for Sensitive Groups']
aqi_counts = df['AQI Category'].value_counts()
plt.pie(aqi_counts, labels=aqi_categories, autopct='%1.1f%%')
plt.title('Proportion of AQI Categories')
plt.axis('equal')
plt.show()
```

```python
# Bar plot: Top 10 countries with the highest AQI
top_10_countries = df.nlargest(10, 'AQI Value')
plt.bar(top_10_countries['Country'], top_10_countries['AQI Value'])
plt.xlabel('Country')
plt.ylabel('AQI Value')
plt.title('Top 10 Countries with the Highest AQI')
plt.xticks(rotation=90)
plt.show()
```

```python
#Histogram
plt.hist(df['PM2.5 AQI Value'], bins=10)
plt.xlabel('PM2.5 AQI Value')
plt.ylabel('Frequency')
plt.title('Distribution of PM2.5 AQI Values')
plt.show()
```

```python
#Box Plot
plt.boxplot([df.loc[df['AQI Category'] == category, 'AQI Value'] for category in df['AQI Category'].unique()])
plt.xlabel('AQI Category')
plt.ylabel('AQI Value')
plt.title('Comparison of AQI Values by AQI Category')
plt.xticks(range(1, len(df['AQI Category'].unique()) + 1), df['AQI Category'].unique())
plt.show()
```

```python
#Stacked Bar Graph
category_counts = df.groupby('Country')['AQI Category'].value_counts().unstack()
category_counts.plot(kind='bar', stacked=True ,figsize=(30, 6))
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Comparison of AQI Categories by City')
plt.legend(title='AQI Category')
plt.xticks(rotation=45)
plt.show()
```

```python
#violin plot
plt.figure(figsize=(8, 6))
plt.violinplot([df.loc[df['AQI Category'] == category, 'AQI Value'] for category in df['AQI Category'].unique()])
plt.xlabel('AQI Category')
plt.ylabel('AQI Value')
plt.title('Distribution of AQI Values by AQI Category')
plt.xticks(range(1, len(df['AQI Category'].unique()) + 1), df['AQI Category'].unique())
plt.show()
```

```python
#HeatMap
corr_matrix = df[['AQI Value', 'CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']].corr()
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation=45)
plt.yticks(range(len(corr_matrix)), corr_matrix.columns)
plt.title('Correlation Matrix of AQI Values')
plt.show()
```

```python
#Geographic
plt.scatter(df['lng'], df['lat'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Cities on a Map')
plt.show()
```

```python
co_aqi_categories = df['CO AQI Category'].value_counts()
labels = co_aqi_categories.index
sizes = co_aqi_categories.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Proportion of AQI Categories based on CO AQI Value')
plt.show()
```

```python
no2_aqi_categories = df['NO2 AQI Category'].value_counts()
labels = no2_aqi_categories.index
sizes = no2_aqi_categories.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Proportion of AQI Categories based on NO2 AQI Value')
plt.show()
```

```python
pm25_aqi_categories = df['PM2.5 AQI Category'].value_counts()
labels = pm25_aqi_categories.index
sizes = pm25_aqi_categories.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Proportion of AQI Categories based on PM2.5 AQI Value')
plt.show()
```

```python
#Density Plot
df['AQI Value'].plot(kind='density')
plt.xlabel('AQI Value')
plt.ylabel('Density')
plt.title('Density Plot of AQI Values')
plt.show()
```

```python
corr_matrix = df[['AQI Value', 'CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']].corr()
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation=45)
plt.yticks(range(len(corr_matrix)), corr_matrix.columns)
plt.title('Correlation Matrix of AQI Values')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='PM2.5 AQI Value', by=['AQI Category', 'City'])
plt.xlabel('AQI Category, City')
plt.ylabel('PM2.5 AQI Value')
plt.title('Comparison of PM2.5 AQI Values by AQI Category and City')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

# Create a new column to categorize AQI Categories and City combinations
df['Category_City'] = df['AQI Category'] + ' - ' + df['Country']

# Create a facet grid plot
g = sns.FacetGrid(df, col='Country', col_wrap=5)
g.map(plt.scatter, 'CO AQI Value', 'Ozone AQI Value')
g.add_legend()
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Relationship between CO AQI Value and Ozone AQI Value by Country')
plt.show()
```
